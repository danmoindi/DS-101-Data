```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (time or angle in radians)
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Compute sine values

# Create the plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="Sine Wave", color="blue")

# Customize the plot
plt.xlabel("X values")
plt.ylabel("Sin(X)")
plt.title("Sine Wave using NumPy and Matplotlib")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
